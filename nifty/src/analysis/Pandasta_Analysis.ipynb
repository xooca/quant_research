```python
import lightning.pytorch as pl
from torch.utils.data import random_split, DataLoader

# Note - you must have torchvision installed for this example
from torchvision.datasets import MNIST
from torchvision import transforms
from typing import List

class TabDataModule(pl.LightningDataModule):
    def __init__(self, 
                 train_df: pd.DataFrame,
                 test_df:pd.DataFrame,
                 val_df: pd.DataFrame,
                 feature_columns: List, 
                 target_column: str):
        super().__init__()
        self.df = df
        self.feature_columns = feature_columns


    def setup(self, stage: str):
        # Assign train/val datasets for use in dataloaders
        if stage == "fit":
            mnist_full = MNIST(self.data_dir, train=True, transform=self.transform)
            self.mnist_train, self.mnist_val = random_split(mnist_full, [55000, 5000])


        # Assign test dataset for use in dataloader(s)
        if stage == "test":
            self.mnist_test = MNIST(self.data_dir, train=False, transform=self.transform)

        if stage == "predict":
            self.mnist_predict = MNIST(self.data_dir, train=False, transform=self.transform)

    def train_dataloader(self):
        return DataLoader(self.mnist_train, batch_size=32)

    def val_dataloader(self):
        return DataLoader(self.mnist_val, batch_size=32)

    def test_dataloader(self):
        return DataLoader(self.mnist_test, batch_size=32)

    def predict_dataloader(self):
        return DataLoader(self.mnist_predict, batch_size=32)
```

```python
import pickle
def get_df(path,alldf=None,recreate_alldf=False,sub_label = 'metric_label__minus_15_1min_close__label_generator_4class_11_1_7',label = None,load_param_label = None):
  param = {}

  lst=[]
  if recreate_alldf:
    dflist = []
    for met in metriclist:
      try:
        with open(f"{path}/{met}", 'rb') as f:
            x = pickle.load(f)
            df = pd.DataFrame(x)
            df['label'] = "_".join(met.split('_')[:-3])
            df['label_sub_cat'] = met.split(".")[0]
            dflist.append(df)
      except Exception as e:
        print(f"{path}/{met} is not loading")
    alldf = pd.concat(dflist)
  if label is None:
    df = alldf[alldf['label_sub_cat']==sub_label][['label','balanced_accuracy_score','roc_auc_score','f1_score','precision_score','recall_score','threshold_probability']]
  else:
    df = alldf[alldf['label']==label][['label','label_sub_cat','balanced_accuracy_score','roc_auc_score','f1_score','precision_score','recall_score','threshold_probability']]
    lst = list(set(df['label_sub_cat'].tolist()))
  if load_param_label is not None:
    with open(f"{path}/bestmodelparam_{load_param_label}.pkl", 'rb') as f:
        param = pickle.load(f)
  return df,param,alldf,lst
```
